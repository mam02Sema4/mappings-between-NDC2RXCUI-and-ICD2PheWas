```python
# map phewas to name
df = pd.read_csv("../raw_data/phewas/phecode_definitions1.2.csv", dtype=str)
df.head()
```

```python
phe2name = dict()

def p2n(x):
    phe2name[x.phecode] = x.phenotype
    
df.apply(p2n, axis=1)

pkl_dump(phe2name, "../resources/phecode2phetype.pkl")
```

```python
# ccs to name

df = pd.read_csv("../raw_data/Single_Level_CCS_2015/prlabel 2014.csv", dtype=str)
df.head()
```

```python
ccs2name = dict()

def c2n(x):
    ccs2name[x['CCS PROCEDURE CATEGORIES']] = x['CCS PROCEDURE CATEGORIES LABELS']
    
df.apply(c2n, axis=1)

pkl_dump(ccs2name, "../resources/ccs2ccsname.pkl")
```
